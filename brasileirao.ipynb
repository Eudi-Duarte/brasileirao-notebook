```python
## Lets set data type for "data" in order to assure ordering in future steps.
## Also, adds a formated date to show date in Brazil format.

partidas["data"] = pd.to_datetime(partidas["data"], dayfirst=True)
partidas["fdata"] = partidas["data"].dt.strftime("%d/%m/%Y")

partidas.info()
```

### Historical aggregated measures
The idea is to retrieve information regarding last N matches of a club.

```python
partidas.sort_values("data", ascending=False).head(10)
```

```python
# Represents the team historic data
class TeamHistory:

    def __init__(self, team_name, n_matches=5, reference_date=date.today()):
        self.team_name = team_name
        self.n_matches = n_matches
        self.reference_date = reference_date
        self.history_as_host = partidas.loc[(partidas["mandante"] == team_name) & (partidas["data"] < reference_date)].sort_values("data", ascending=False).head(n_matches)
        self.history_as_away = partidas.loc[(partidas["visitante"] == team_name) & (partidas["data"] < reference_date)].sort_values("data", ascending=False).head(n_matches)
        self.history = pd.concat([self.history_as_host, self.history_as_away], ignore_index=True).sort_values("data", ascending=False)
        self.gols_as_host = int(self.history_as_host["mandante_Placar"].sum())
        self.gols_as_away = int(self.history_as_away["visitante_Placar"].sum())
        self.avg_gols_as_host = self.gols_as_host / n_matches
        self.avg_gols_as_away = self.gols_as_away / n_matches


## Function that return all matches of a given date
def get_matches(date=date.today()):
    return partidas.loc[partidas["data"] == date][["rodata", "data", "fdata", "hora", "mandante", "visitante", "vencedor", "mandante_Placar", "visitante_Placar"]]
    
```

### Get all matches of a day

In future steps, we will code a logic for define what matches should be shown currently. As now we don't have current data, lets use the last stored date.

```python
## Defines which matches will be included in the home page

last_round = partidas["data"].max() 
brasileirao_matches = get_matches(last_round)
brasileirao_matches
```

```python
## Store it to csv

brasileirao_matches.to_csv('data/clean/brasileirao_matches.csv')
```

```python
## Store last macthes dataframe for each time in the next round

teams = set(pd.concat([brasileirao_matches["mandante"], brasileirao_matches["visitante"]], ignore_index=True))

for team in teams:
    team_history = TeamHistory(team_name=team, reference_date=last_round)
    team_history.history.to_csv(f"data/clean/teams/{team.lower()}.csv")
```
