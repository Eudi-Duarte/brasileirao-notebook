### Historical aggregated measures
The idea is to retrieve information regarding last N matches of a club.

```python
# get team last N macthes as host
def get_host_history(team_name, date=datetime.now(), n_matches=10):
    return partidas.loc[(partidas["mandante"] == team_name) & (partidas["data"] < date)].sort_values("data", ascending=False).head(n_matches)

## Create a function that return agg values for the last N matches of a club as host, given a date (default=current_date)
def get_host_history_agg(team_name, date=datetime.now(), n_matches=10):
    history_df = partidas.loc[(partidas["mandante"] == team_name) & (partidas["data"] < date)].sort_values("data", ascending=False).head(n_matches)

    json_agg = {
        "team_name":team_name,
        "gols_as_host":history_df.loc[history_df["mandante"] == team_name]["mandante_Placar"].apply(sum, by_row=False),
        "gols_as_away":history_df.loc[history_df["visitante"] == team_name]["visitante_Placar"].apply(sum, by_row=False),
    }
    
    return pd.Series(json_agg)

## Function that return all matches of a given date
def get_matches(date=date.today()):
    return partidas.loc[partidas["data"] == date][["rodata", "data", "hora", "mandante", "visitante", "vencedor", "mandante_Placar", "visitante_Placar"]]
    
```
